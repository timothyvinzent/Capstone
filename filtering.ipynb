```python
# now we need to create a function that will filter the dataframes by closest latitude and longitude values to the user's location
# we will use the haversine formula to calculate the distance between the user's location and the location of the data points
# to increase efficiency we know that all the data points we have are accurate to two decimal points, therefore we generate a list of 5 possible combinations which would be closest to the users location 
# using the 5 possible coordinates we filter the dataframes and return the filtered dataframes, if there are no datapoint for that location from the list we take the next values

# generating a list of coordinates closest to the user's location using the haversine formula
u_lat = 46.8182
u_lon = 8.2275

def harversine (lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2*np.arcsin(np.sqrt(a))
    return R*c


def reverse_harversine(lat, lon):
    R = 6371
    lat, lon = map(np.radians, [lat, lon])
    latitudes = np.arange(lat - 0.05, lat + 0.06, 0.01)
    longitudes = np.arange(lon - 0.05, lon + 0.06, 0.01)
    lats_longs = [(round(lat_,2), round(lon_,2)) for lat_ in latitudes for lon_ in longitudes]
    return lats_longs[:5]

print(reverse_harversine(u_lat, u_lon))
```

```python
def check_point_within_range(u_lat, u_lon, lat_range, long_range):
    if lat_range[0] <= u_lat <= lat_range[1] and long_range[0] <= u_lon <= long_range[1]:
        return True
    else:
        return False
```

```python
def plot_creator(lat, lon, df, column_name, title, y_label):
    switzerland_lat_range = (45.8172, 47.8087)
    switzerland_long_range = (5.9569, 10.4922)

    france_lat_range = (41.3712, 51.0918)
    france_long_range = (-5.1428, 9.5600)

    rwanda_lat_range = (-2.8400, -1.0471)
    rwanda_long_range = (28.8614, 30.8958)

    u_lat = lat
    u_lon = lon

    country_loc = [(switzerland_lat_range, switzerland_long_range), (france_lat_range, france_long_range), (rwanda_lat_range, rwanda_long_range)]
    countries = ['Switzerland', 'France', 'Rwanda']

    checkpoint = False
    your_country= []

    for i in range(len(country_loc)):

        if check_point_within_range(u_lat, u_lon, country_loc[i][0], country_loc[i][1]) == True:
            
            checkpoint = True
            your_country.append(countries[i])
            continue
        else:
            continue
    if checkpoint == False:
        print("Your location is not within the range of the data we have")
    else:
        print(f"Your location is within the range of {your_country[0]}")
        df_filtered = df[(df['lon'].round(1) == round(u_lon, 1)) & (df['lat'].round(1) == round(u_lat, 1))]
        #df_filtered['Temperature_Air_2m_Max_24h'] = df_filtered['Temperature_Air_2m_Max_24h'].apply(lambda x: x - 273.15)
        df_filtered['time'] = pd.to_datetime(df_filtered['time'], format='%Y-%m-%d %H:%M:%S')
        plt.figure(figsize=(20,10))
        plt.plot(df_filtered['time'], df_filtered[column_name])
        plt.title(title)
        plt.xticks(df_filtered['time'][::90], df_filtered['time'][::90].dt.strftime('%Y-%m'), rotation=90)
        plt.ylabel(y_label)
        plt.show()
        return df_filtered
    
    
```
